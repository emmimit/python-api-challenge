## Generate Cities List

```python
lat_lngs = []
cities = []


lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = list(zip(lats,lngs))

for lat_lngs in lat_lngs:
    city = citipy.nearest_city(lat_lngs[0], lat_lngs[1]).city_name

    if city not in cities:
        cities.append(city)

len(cities)
```

## Linear Regression

```python
n_hem = city_weather_df.loc[city_weather_df["lat"] >=0, :]
s_hem = city_weather_df.loc[city_weather_df["lat"] < 0, :] 
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = n_hem["lat"]
y_values = n_hem["temp_max"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept, 2))

plt.scatter(x_values, y_values, facecolor="lightgreen", edgecolor="black")
plt.plot(x_values, regress_values, "r-")
plt.title("Northern Hemisphere - Max Temp vs Latitude")
plt.xlabel('Latitude')
plt.ylabel('Max Temperature')
plt.annotate(line_eq, (0,0), fontsize=15, color="red")

print(f"The r-value is:{rvalue**2}")
plt.savefig(" North Hem - Max Temp vs Latitude")
```
